```r
x <- rnorm(100)
```

```r
y <- x + rnorm(100,sd=0.5)
```

```r
plot(x,y)
```

```r
lmout <- lm(y~x)
```

```r
print(lmout)
```

```r
scatter.smooth(x, y, main="x ~ y")
```

```r
cor(y, x)
```

```r
linearMod <- lm(x ~ y)
```

```r
print(linearMod)
```

```r
summary(linearMod)
```

```r
AIC(linearMod)
```

```r
BIC(linearMod)
```
